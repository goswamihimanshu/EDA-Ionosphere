## Feature Engineering

## Creating Independent and Dependent Variables

```python
# Independent variables
X = df.drop(columns = ['Class'],axis = 1)
X.head()
```

```python
# Dependent variable
y = df['Class']
y.head()
```

## Spliting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Scaling the numerical features:

```python
# Scaling of data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log = LogisticRegression()
```

```python
model_log.fit(X_train,y_train)
```

```python
y_pred_log = model_log.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test,y_pred_log)
plt.figure(figsize = (5,3))

sns.heatmap(cm,annot = True,fmt = '0.2f',xticklabels=['class-0','class-1'],
           yticklabels=['class-0','class-1'],cbar=False)
plt.xlabel('Predicted Values')
plt.ylabel('True Values')
```

```python
cr = classification_report(y_test,y_pred_log)
print(cr)
```

```python
log_model_score = model_log.score(X_test,y_test)
print(log_model_score)
```
